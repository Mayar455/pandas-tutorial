### Importing Libraries

```
import numpy as np
import pandas as pd
```

```
citiesDataFrame = pd.read_csv("cities.csv")
```

```
citiesDataFrame.head()
```

```
citiesDataFrame.columns = citiesDataFrame.columns.str.replace(' ','')
citiesDataFrame.columns = citiesDataFrame.columns.str.replace('"','')
```

```
citiesDataFrame.head()
```

### INSERT

```
data = {'LatD' : 39,
        'LatM' : 57,
        'LatS' : 23,
        'NS' : '"N"',
        'LonD' : 41,
        'LonM' : 10,
        'LonS' : 12,
        'EW' : '"E"',
        'City' : '"Erzurum"',
        'State' : "TUR"}
```

```
citiesDataFrame = citiesDataFrame.append(data, ignore_index=True)
```

```
citiesDataFrame.tail()
```

### UPDATE

```
citiesDataFrame.loc[citiesDataFrame['City'] == '"Erzurum"', 'State'] = "ERZ"
```

```
citiesDataFrame.tail()
```

### DELETE

```
citiesDataFrame = citiesDataFrame[citiesDataFrame['City'] != '"Erzurum"']
```
